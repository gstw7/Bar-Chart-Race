```python
import pandas as pd
import bar_chart_race as bcr

pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_excel('covid-br.xlsx')
df.head()
```

```python
def df_auxiliar(df):
    
    df_aux = pd.DataFrame({'colunas' : df.columns,
                        'tipo': df.dtypes,
                        'missing' : df.isna().sum(),
                        'size' : df.shape[0],
                        'unicos': df.nunique()})
    df_aux['percentual % '] = round(df_aux['missing'] / df_aux['size'],4)*100
    return df_aux
```

```python
df_aux = df_auxiliar(df)
df_aux
```

```python
pct_null = df.isnull().sum() / len(df)
missing_features = pct_null[pct_null > 0.15].index
df.drop(missing_features, axis=1, inplace=True)
```

```python
df.head()
```

```python
df = df[df['estado'] == 'AL']
df.dropna(inplace=True)
df.head()
```

```python
df_al_bcr= pd.DataFrame(df.groupby(['municipio', 'data']).sum()['casosAcumulado']).unstack().T.droplevel(level=0)
df_al_bcr.head()
```

```python
bcr.bar_chart_race(df_al_bcr, n_bars=10, filter_column_colors=True, title='Casos Acumulados do COVID-19 em Alagoas')
```
